```python
import pandas as pd
```

```python
df= pd.read_csv("premier-league-tables.csv", index_col=False)
df2 = df.rename(index=str, columns={'Season_End_Year':"Year","Rk":"Ranking", "MP":"Matches_Played", "GF":"Goals_Scored","GA":"Goals_Conceded", "GD":"Goal_Difference", "Pts":"Points"})
```

```python
df2 = df2[df2['Year'] >= 2010]
```

```python
missing_values_count = df2.isna().sum()
```

```python
missing = pd.concat([df2.isnull().sum(), 100 * df2.isnull().mean()], axis=1)
missing.columns = ['count', '%']
missing_sorted = missing.sort_values(by='count')
```

```python
missing_sorted
```

```python
df2= df2.fillna("Nothing")
#This shows that the team did not get qualify another League or get relegated into the league below so 
#we give Nan the value of nothing
```

```python
df2.isnull().sum()
```

```python
team_2023 = df2[(df2['Year'] == 2023) & (~df2["Notes"].str.contains("Relegated"))]
#This was done to select the teams from 2023 that stayed in Premier League in the year 2024
```

```python
team_2023
```

```python
prem_2024=df2[df2['Team'].isin(team_2023["Team"])]
```

```python
prem_2024
```

```python
prem_2024["Team"].value_counts()
```

```python
max_gs= prem_2024.groupby('Team')['Goals_Scored'].sum()
max_gs.sort_values(ascending=False)
#Amount Teams have scored since 2010
```

```python
avg_rank=prem_2024.groupby("Team")["Ranking"].mean()
avg_rank.sort_values(ascending=True)
#Average ranking for teams since 2010/The higher the better
```

```python
conceded_goals=prem_2024.groupby("Team")["Goals_Conceded"].sum()
conceded_goals.sort_values(ascending=True)
#A factor that plays into this is how long the teams have played in the Premier League
```

```python
conceded_goals_gp=prem_2024.groupby("Team")["Goals_Conceded"].sum()/prem_2024.groupby("Team")["Matches_Played"].sum()
conceded_goals_gp.sort_values(ascending=True)
```

```python
team_stats = prem_2024.groupby('Team').agg({'Matches_Played': 'sum', 'Goals_Conceded': 'sum'})
team_stats
```
